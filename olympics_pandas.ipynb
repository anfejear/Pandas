## Importación de librerías

En este trabajo se emplean las librerías ``panda`` y ``numpy``.

```python
import pandas as pd 
import numpy as np 
```

## Importación de datos

Se importan los _datasets_ de juegos olímpicos, población y producto interno bruto (PIB) como _dataframes_ de ``panda`` con las etiquetas ``df_OG``, ``df_POP`` y ``df_GDP``, respectivamente.

```python
df_OG = pd.read_csv('olympics.csv', index_col = 0, skiprows=1)
df_POP = pd.read_csv('POP/TOTL.csv', index_col = 0, skiprows=4)
df_POP=df_POP.drop('Country Code',axis=1) # Se elimina la columna Contry Code
df_POP=df_POP.drop('Indicator Code',axis=1)  # Se elimina la columna Indicator Code
df_POP=df_POP.drop('Indicator Name',axis=1)  # Se elimina la columna Indicator Name
df_POP=df_POP.drop('Unnamed: 62',axis=1)  # Se elimina la columna Unnamed: 62
df_GDP = pd.read_csv('GDP/GDP.csv', index_col = 0, skiprows=4)
df_GDP=df_GDP.drop('Country Code',axis=1)  # Se elimina la columna Contry Code
df_GDP=df_GDP.drop('Indicator Code',axis=1)  # Se elimina la columna Indicator Code
df_GDP=df_GDP.drop('Indicator Name',axis=1)  # Se elimina la columna Indicator Name
df_GDP=df_GDP.drop('Unnamed: 62',axis=1)  # Se elimina la columna Unnamed: 62
```

En este caso, se etiquetan las columnas ``01``, ``02`` y ``03`` del _dataframe_ juegos olímpicos ``df_OG`` con los nombres de las medallas **Gold**, **Silver** y **Bronze**, respectivamente,

```python
for col in df_OG.columns:
    if col[:2]=='01':
        df_OG.rename(columns={col:'Gold' + col[4:]}, inplace=True)
    if col[:2]=='02':
        df_OG.rename(columns={col:'Silver' + col[4:]}, inplace=True)
    if col[:2]=='03':
        df_OG.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
    if col[:1]=='№':
        df_OG.rename(columns={col:'#' + col[1:]}, inplace=True) 
df_OG.head()
```

Con el fin de trabajar con índeces homogeneos en los tres _dataframes_, se suprime el código del país en el ``index`` del dataframe ``df_OG``.

```python
index0=df_OG.index
for ind in np.arange(len(index0)):
    df_OG.rename(index={index0[ind]: index0[ind][:index0[ind].find("(")-1]},inplace=True)
df_OG.head()
```

## Promedio de medallas

Con el fin de calcular un promedio ponderado de las medallas según el país y el juego olímpico en el cual participó (verano o invierno), se asigna un peso a cada medalla, es decir que, las medallas oro, plata y bronce  tienen un peso de 3, 2 y 1, respectivamente.  

```python
Summer=(df_OG['Gold']*3+df_OG['Silver']*2+df_OG['Bronze'])/df_OG['Total'];
Winter=(df_OG['Gold.1']*3+df_OG['Silver.1']*2+df_OG['Bronze.1'])/df_OG['Total.1'];
Total=(df_OG['Gold.2']*3+df_OG['Silver.2']*2+df_OG['Bronze.2'])/df_OG['Combined total'];
```

Se realiza un ``merge()`` para trabajar los datos de los promedios en un mismo _dataframe_

```python
df_Olimpic=pd.merge(pd.DataFrame(Summer),pd.DataFrame(Winter),how='outer',left_index=True,right_index=True);
df_Olimpic=pd.merge(df_Olimpic,pd.DataFrame(Total),how='outer',left_index=True,right_index=True);
df_Olimpic=pd.merge(df_Olimpic,pd.DataFrame(df_OG['# Summer']),how='outer',left_index=True,right_index=True);
df_Olimpic=pd.merge(df_Olimpic,pd.DataFrame(df_OG['# Winter']),how='outer',left_index=True,right_index=True);
df_Olimpic=pd.merge(df_Olimpic,pd.DataFrame(df_OG['# Games']),how='outer',left_index=True,right_index=True);
df_Olimpic.rename(columns={'0_x':'Mean Summer'},inplace=True);
df_Olimpic.rename(columns={'0_y':'Mean Winter'},inplace=True);
df_Olimpic.rename(columns={0:'Mean Total Games'},inplace=True);
df_Olimpic.head()
```

## Promedio de la población mundial y el PIB

Se calcula el promedio de la población y el PIB desde 1960 hasta 2017 en cada país.

```python
indexPOP=df_POP.index;
indexGDP=df_GDP.index;
indexOG=df_OG.index;
S_GDP=df_GDP.loc[indexGDP].mean(axis=1);
S_POP=df_GDP.loc[indexPOP].mean(axis=1);
S_GDP=pd.DataFrame(S_GDP);
S_POP=pd.DataFrame(S_POP);
S_GDP.rename(columns={0:'Mean GDP'},inplace=True);
S_POP.rename(columns={0:'Mean POP'},inplace=True);
```

Ahora, se realiza un ``merge()`` de los promedios de las medallas con los promedios de la población y el PIB.

```python
df_group=pd.merge(df_Olimpic,S_GDP,how='inner',left_index=True,right_index=True);
df_group=pd.merge(df_group,S_POP,how='inner',left_index=True,right_index=True)
df_group.head()
```

## Análisis de datos

### 1. Análisis de datos sin incluir la población y el PIB

En primer lugar, se realiza un análisis teniendo en cuenta solo las medallas ganadas en los juegos olímpicos de invierno y verano según el número de juegos jugados en cada país. Además, se analiza el total de medallas ganadas en estos juegos según el país.

```python
weight_Winter=df_Olimpic['Mean Winter'][df_Olimpic['# Winter']!=0]/df_Olimpic['# Winter'][df_Olimpic['# Winter']!=0];
weight_Summer=df_Olimpic['Mean Summer'][df_Olimpic['# Summer']!=0]/df_Olimpic['# Summer'][df_Olimpic['# Summer']!=0];
weight_Total=df_Olimpic['Mean Total Games'][df_Olimpic['# Games']!=0]/df_Olimpic['# Games'][df_Olimpic['# Games']!=0];
```

Se eliminan los campos que están vacíos, es decir, con valores ``NaN``, mediante la función ``dropna()``.

```python
weight_Winter=weight_Winter.dropna(how='any');
weight_Summer=weight_Summer.dropna(how='any');
weight_Total=weight_Total.dropna(how='any');
```

Se organiza de mayor a menor los países de acuerdo a su peso asignado.

```python
Max_Winter=weight_Winter.sort_values(ascending=False).index[0];
Max_Summer=weight_Summer.sort_values(ascending=False).index[0];
Max_Total=weight_Total.sort_values(ascending=False).index[0];
```

### 1. Análisis de datos incluyendo la población y el PIB

En este caso, se realiza un análisis teniendo en cuenta la población y el PIB.

```python
weight_Winter=df_group['Mean Winter'][df_group['# Winter']!=0]/df_group['# Winter'][df_group['# Winter']!=0];
weight_Summer=df_group['Mean Summer'][df_group['# Summer']!=0]/df_group['# Summer'][df_group['# Summer']!=0];
weight_Total=df_group['Mean Total Games'][df_group['# Games']!=0]/df_group['# Games'][df_group['# Games']!=0];
wight_Winter_GDP=weight_Winter/df_group['Mean GDP'];
weight_Summer_GDP=weight_Summer/df_group['Mean GDP'];
weight_Total_GDP=weight_Total/df_group['Mean GDP'];
wight_Winter_POP=weight_Winter/df_group['Mean POP'];
weight_Summer_POP=weight_Summer/df_group['Mean POP'];
weight_Total_POP=weight_Total/df_group['Mean POP'];
WinterT=wight_Winter_GDP+wight_Winter_POP;
SummerT=weight_Summer_GDP+weight_Summer_POP;
TotalT=weight_Total_GDP+weight_Total_POP;
```

Se eliminan los campos que están vacíos, es decir, con valores ``NaN``, mediante la función ``dropna()``.

```python
WinterT=WinterT.dropna(how='any');
SummerT=SummerT.dropna(how='any');
TotalT=TotalT.dropna(how='any');
```

Se organiza de mayor a menor los países de acuerdo a su peso asignado.

```python
Max_Winter=WinterT.sort_values(ascending=False).index[0];
Max_Summer=SummerT.sort_values(ascending=False).index[0];
Max_Total=TotalT.sort_values(ascending=False).index[0];
```
