```python
for col in olympics_data.columns:
    if col[:2]=='01':
        olympics_data.rename(columns={col:'Gold' + col[4:]}, inplace=True)
    if col[:2]=='02':
        olympics_data.rename(columns={col:'Silver' + col[4:]}, inplace=True)
    if col[:2]=='03':
        olympics_data.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
    if col[:1]=='№':
        olympics_data.rename(columns={col:'#' + col[1:]}, inplace=True) 

olympics_data.head()
```
